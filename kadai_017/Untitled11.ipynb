```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")

humans = [
    Human("Taro", 25),
    Human("Hanako", 17),
    Human("Jiro", 20)
]

for human in humans:
    human.check_adult()
```
